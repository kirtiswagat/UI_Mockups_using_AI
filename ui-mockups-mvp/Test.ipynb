```python
from openai import OpenAI
import os, base64
from dotenv import load_dotenv
from IPython.display import Image, display

# 1ï¸âƒ£ Load your .env file (if you have one)
load_dotenv()

# 2ï¸âƒ£ Initialize client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 3ï¸âƒ£ Define your prompt
prompt = "A clean, professional dashboard UI for a portfolio management app, light theme, modern cards, top navigation bar, and summary metrics"

# 4ï¸âƒ£ Call OpenAI Image API
print("ðŸ”„ Generating UI Mockup, please wait...")
response = client.images.generate(
    model="gpt-image-1",
    prompt=prompt,
    n=1,
    size="1024x1024"
)

# 5ï¸âƒ£ Decode and save image
b64 = response.data[0].b64_json
image_bytes = base64.b64decode(b64)

output_file = "test_mockup.png"
with open(output_file, "wb") as f:
    f.write(image_bytes)

print(f"âœ… Image saved as {output_file}")

# 6ï¸âƒ£ Display inline in Jupyter
display(Image(filename=output_file))
```
